```python
import cv2
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_n = cv2.imread('A-PF-Apr-002-2017-1.jpg',1)
imgray = cv2.cvtColor(img_n,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,0)
image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
img_c=img_n.copy()
```

```python
print (len(contours))
cnt = contours[0]
```

```python
for c in contours:
    x,y,w,h=cv2.boundingRect(c)
    cv2.rectangle(img_c,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
plt.figure(figsize=(40,20))
plt.imshow(img_c,'gray')
```

```python
img_a=img_n.copy()
for c in contours:
    accuracy=0.1*cv2.arcLength(c,True)
    approx=cv2.approxPolyDP(c,accuracy,True)
    cv2.drawContours(img_a,[approx],0,[0,255,0],2)
```

```python
plt.figure(figsize=(40,20))
plt.imshow(img_a,'gray')
```

```python
img_h=img_n.copy()
for c in contours:
    hull=cv2.convexHull(c)
    cv2.drawContours(img_h,[hull],0,(255,0,0),2)
    
```

```python
plt.figure(figsize=(40,20))
plt.imshow(img_h,'gray')
```

```python
img_hu=img_n.copy()
hull=[]
for c in contours:
    hul=cv2.convexHull(c)
    hull.append(hul)
    cv2.drawContours(img_hu,[hul],0,(255,0,0),2)
```

```python
plt.figure(figsize=(40,20))
plt.imshow(img_hu,'gray')
```

```python
img_roi=img_n.copy()
kernel = np.ones((7,8), np.uint8)
ret,thresh = cv2.threshold(imgray,127,255,cv2.THRESH_BINARY_INV)
img_dilation = cv2.dilate(thresh, kernel, iterations=1)
im2,ctrs, hier = cv2.findContours(img_dilation.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

#sorted_ctrs = sorted(ctrs, key=lambda ctr: cv2.boundingRect(ctr)[0])
for i, ctr in enumerate(ctrs):
    # Get bounding box
    x, y, w, h = cv2.boundingRect(ctr)
 
    # Getting ROI
    roi = image[y:y+h, x:x+w]
 
    # show ROI
    #cv2.imshow('segment no:'+str(i),roi)
    cv2.rectangle(img_roi,(x,y),( x + w, y + h ),(0,255,0),2)
    
    
```

```python
plt.figure(figsize=(40,20))
plt.imshow(img_roi,'gray')
```
