```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('A-PF-Apr-002-2017-1.jpg',1)
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(20,10))
ret,thresh1=cv2.threshold(hsv,127,255,cv2.THRESH_BINARY)

plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(thresh1,'gray')

```

```python
plt.figure(figsize=(20,10))
et2,tho = cv2.threshold(hsv,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(tho,'gray')
```

```python
plt.figure(figsize=(20,10))
th2 = cv2.adaptiveThreshold(hsv,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,7,2)
plt.imshow(th2,'gray')
```

```python
cv2.imwrite('simple_thresholding.png',thresh1)
cv2.imwrite('adaptive_thresholding.png',th2)
cv2.imwrite('otsu_thresholding.png',tho)
```

```python
plt.figure(figsize=(20,10))
blur = cv2.GaussianBlur(img,(5,5),0)
plt.imshow(blur,'gray')
```

```python
cv2.imwrite('gaussian_blur.png',blur)
```

```python
mser = cv2.MSER_create()
img1=cv2.imread('A-PF-Apr-002-2017-1.jpg',1)
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
vis = img.copy()
regions, _ = mser.detectRegions(gray)

hulls = [cv2.convexHull(p.reshape(-1, 1, 2)) for p in regions]

cv2.polylines(vis, hulls, 1, (0, 255, 0))
plt.figure(figsize=(40,20))
plt.imshow(vis)
```

```python
for corr in regions:
    bbox=cv2.boundingRect(corr)
    x,y,w,h=bbox
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
plt.figure(figsize=(40,20))
plt.imshow(img)
```
